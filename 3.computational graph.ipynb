```python
import torch
```

```python
w = torch.tensor([1.], requires_grad = True)
x = torch.tensor([2.], requires_grad = True)
```

```python
a = torch.add(w, x)
a.retain_grad() # The grad of a is stored
b = torch.add(w, 1)
y = torch.mul(a, b)
```

```python
y.backward()
```

```python
print(w.grad, a.grad, b.grad, y.grad) # The grad of non_leaf node except a has been freed 
```

```python
print("w:{} {}\nx:{} {}\na:{} {}\nb:{} {}\ny:{} {}".format(w, w.is_leaf, x, x.is_leaf, a, a.is_leaf, b, b.is_leaf, y, y.is_leaf))
```
